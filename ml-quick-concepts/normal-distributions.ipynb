```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/noahgift/\
regression-concepts/master/height-weight-25k.csv")
```

```python
print(df.describe())
```

```python
sns.lmplot(x="Height-Inches", y="Weight-Pounds", data=df)
```

One of the best ways to visualize height and weights normal bell-shaped distribution
is to use a Kernel Density plot:

```python
sns.jointplot(x="Height-Inches", y="Weight-Pounds", data=df, kind="kde");
```
